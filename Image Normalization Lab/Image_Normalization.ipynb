```python
import cv2
import numpy as np
```

```python
image = cv2.imread("input.jpg")

# Convert to grayscale (8-bit)

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Resize to 70% using Linear Interpolation

scale = 0.7

new_width = int(gray.shape[1] * scale)
new_height = int(gray.shape[0] * scale)

resized = cv2.resize(
    gray,
    (new_width, new_height),
    interpolation=cv2.INTER_LINEAR
)

restored = cv2.resize(
    resized,
    (gray.shape[1], gray.shape[0]),
    interpolation=cv2.INTER_LINEAR
)
```

```python
cv2.imwrite("gray_image.png", gray)
cv2.imwrite("resized_07.png", resized)
cv2.imwrite("restored.png", restored)
```

```python
# Convert images to float to avoid overflow
original = gray.astype(np.float32)
reconstructed = restored.astype(np.float32)

# Difference
difference = original - reconstructed

# Square the differences
squared_difference = difference ** 2

# Average of squared differences (MSE)
average_squared_difference = np.mean(squared_difference)

print("Average Squared Difference (MSE):", average_squared_difference)
```
